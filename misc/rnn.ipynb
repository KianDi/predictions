```python
import torch

import torch.nn as nn
```

```python
rnn = nn.RNN(10, 20, 2)
input = torch.randn(5, 3, 10)
h0 = torch.randn(2, 3, 20)
output, hn = rnn(input, h0)
output, hn
```

```python
rnn = nn.RNN(input_size=10, hidden_size=20, num_layers=2)
input = torch.randn(5, 3, 10)
input
```

```python
h0 = torch.randn(2, 3, 20)
h0
```

```python
output, hn = rnn(input, h0)
output, hn
```
